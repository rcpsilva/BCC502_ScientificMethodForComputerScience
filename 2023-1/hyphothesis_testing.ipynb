```python
alg1 = np.random.rand(30)
alg2 = np.random.rand(30) 
```

```python
print(np.mean(alg1))
print(np.mean(alg2))
```

```python
print(f'{np.mean(alg1)} +- {np.std(alg1)}')
print(f'{np.mean(alg2)} +- {np.std(alg2)}')
```

```python
from tqdm import tqdm
n = 1000000
malg1 = []
malg2 = []


for i in tqdm(range(n)):
    alg1 = np.random.rand(30)
    alg2 = np.random.rand(30)
    malg1.append(np.mean(alg1))
    malg2.append(np.mean(alg2))
```

```python
import matplotlib.pyplot as plt

malg1 = np.array(malg1)
malg2 = np.array(malg2)

plt.hist(malg1-malg2,bins=50)
```
