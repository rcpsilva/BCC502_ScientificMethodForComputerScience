```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([1,2,3,4,5,6,7])
y = (x-4)**2+np.random.randn(7)
plt.plot(x,y,'ro')
```

```python
degree=2
polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(x.reshape(-1,1),y)
x_seq = np.arange(0,8,0.05)
plt.plot(x_seq,polyreg.predict(x_seq.reshape(-1,1)),color="black")
plt.plot(x,y,'ro')
```

```python
x = np.append([4.5],x)
y = np.append([15],y)
plt.plot(x,y,'ro')
plt.plot(x[0],y[0],'bd')
```

```python
degree=2
polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(x.reshape(-1,1),y)
x_seq = np.arange(0,8,0.05)
plt.plot(x_seq,polyreg.predict(x_seq.reshape(-1,1)),color="black")
plt.plot(x,y,'ro')
plt.plot(x[0],y[0],'bd')
```
